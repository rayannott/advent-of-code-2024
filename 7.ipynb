```python
import pathlib
import itertools
import operator
```

```python
DIR = pathlib.Path("inputs")
INPUT = DIR / "7.txt"

DATA = INPUT.read_text()
```

```python
OP_MAP = {
    0: operator.mul,
    1: operator.add,
}
```

```python
eqs = [line.split(':') for line in DATA.split("\n")]
eqs = [(int(eq[0]), list(map(int, eq[1].strip().split()))) for eq in eqs]
```

```python
eqs
```

```python
valid = []

for ans, seq in eqs:
    for ops in itertools.product((0, 1), repeat=len(seq) - 1):
        res = seq[0]
        for val, op in zip(seq[1:], ops):
            res = OP_MAP[op](res, val)
        if res == ans:
            valid.append(res)
            break
```

```python
sum(valid)
```

```python
OP_MAP[2] = lambda x, y: int(str(x) + str(y))
```

```python
valid = []

for ans, seq in eqs:
    for ops in itertools.product((0, 1, 2), repeat=len(seq) - 1):
        res = seq[0]
        for val, op in zip(seq[1:], ops):
            res = OP_MAP[op](res, val)
        if res == ans:
            valid.append(res)
            break
```

```python
sum(valid)
```
